```python
def numpy_boundary_traversal(matrix) :  # function
  top = matrix[0, :]
  right = matrix[1:, -1]
  bottom = matrix[-1, -2 ::-1]
  left = matrix[-2:1:-1, 0]

  return np.concatenate([top, right, bottom, left])


res = numpy_boundary_traversal(a)
print(res)
```

```python
def numpy_alternate_sort(matrix) :
  sort_arr = np.sort(matrix)
  n = matrix.shape[0]
  res = np.empty(n, dtype= matrix.dtype)
  res[::2] = sort_arr[:(n+1)//2]
  res[1::2] = sort_arr[n-1:(n-1)//2:-1]
  return res

res = numpy_alternate_sort(b)
print(res)
```

# Problem 3

```python
import pandas as pd

# creating the data frame from a dict
names = ['Raj', 'Ash', 'Virat', 'Neha', 'Shreya', 'Rahul', 'Bob',
         'Robin', 'Steve', 'Dude']
subjects = ['Math', 'Math', 'Physics', 'English', 'History', 'Physics',
            'Chemistry', 'Physics', 'Chemistry', 'History']

scores = np.random.randint(50, 100, (10,))

dict = {'Name' : names, 'Subject' : subjects,
        'Score' : scores}

df = pd.DataFrame(dict)
df
```

```python
def grade(score) :   # grade function
  if score >= 90 :
    return 'A'
  elif score >= 80 :
    return 'B'
  elif score >= 70 :
    return 'C'
  elif score >= 60 :
    return 'D'
  else :
    return 'F'
```

```python
df['Grade'] = [grade(score) for score in scores]
df
```

```python
df = df.sort_values('Score', ascending=False) # sorting the values
df.reset_index(drop=True, inplace=True)
df
```

```python
df.groupby('Subject')['Score'].mean() # average scores for each subject
```

```python
def pandas_filter_pass(dataframe) :   # filter function
  return dataframe[(dataframe['Grade'] == 'A') | (dataframe['Grade'] == 'B')]
```

```python
new_df = pandas_filter_pass(df)
new_df
```

# Problem 4

```python
import sklearn
import json
```

```python
class Review:
  def __init__(self, text, score) :
    self.text = text
    self.score = score

  def sentiment(self) :
    if self.score <= 2 :
      return 'NEGATIVE'
    elif self.score >= 4 :
      return 'POSITIVE'
    else :
      return 'NEUTRAL'

  def goodbad(self) :
    if self.score >= 4 :
      return 'BAD'
    elif self.score <= 2 :
      return 'GOOD'

# found this dataset online (which contains reviews of users on amazon and their ratings)
file_name = '/content/Books_small.json'
reviews = []
with open(file_name) as f:
  for line in f:
    review = json.loads(line)
    reviews.append(Review(review['reviewText'], review['overall']))
len(reviews)
```

```python
pos_sample = []
neg_sample = []
for review in reviews :
  if review.sentiment() == 'POSITIVE' :
    pos_sample.append(review)
  elif review.sentiment() == 'NEGATIVE' :
    neg_sample.append(review)
```

```python
# getting the required dataset
from random import shuffle
pos_sample = pos_sample[:50]
neg_sample = neg_sample[:50]
samples = pos_sample + neg_sample
shuffle(samples)
len(samples)
```

```python
df = pd.DataFrame([x.text for x in samples], columns=['text'])
df['sentiment'] = [x.sentiment() for x in samples]
df['goodbad'] = [x.goodbad() for x in samples]
df   # this is the dataset, can be used for both 4th and 5th question
```

```python
# dataframe for 4th question
movie_df = df.drop('goodbad', axis=1)
movie_df
```

```python
# feature extraction
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(movie_df, test_size=0.2)
vectorizer = CountVectorizer(max_features=500,lowercase=True ,stop_words='english')

train_x = vectorizer.fit_transform(train_df['text'])
train_y = train_df['sentiment']
test_x = vectorizer.transform(test_df['text'])
test_y = test_df['sentiment']
```

```python
from sklearn.naive_bayes import MultinomialNB  # model training
classifier = MultinomialNB()
classifier.fit(train_x.toarray(), train_y)
y_pred = classifier.predict(test_x.toarray())
# print(y_pred)
```

```python
from sklearn.metrics import accuracy_score  # accuracy
print(accuracy_score(test_y, y_pred))
```

```python
def predict_review_sentiment(model, vectorizer, review) : # function
  review = vectorizer.transform([review])
  return model.predict(review)
```

```python
review = movie_df['text'][0]
print(review)
print(predict_review_sentiment(classifier, vectorizer, review))
```

```python
# dataframe for the 5th question
text_df = df.drop('sentiment', axis = 1)
text_df
```

```python
# feature extraction
from sklearn.feature_extraction.text import TfidfVectorizer

train, test = train_test_split(text_df, test_size =0.25)
vectorizer = TfidfVectorizer(max_features=300,lowercase=True ,stop_words='english')

train_x = vectorizer.fit_transform(train['text'])
train_y = train['goodbad']
test_x = vectorizer.transform(test['text'])
test_y = test['goodbad']
```

```python
# model training and prediction
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(train_x.toarray(), train_y)
y_pred = logistic.predict(test_x.toarray())
```

```python
# evaluation
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
print("Accuracy : ", accuracy_score(test_y, y_pred))
print("Recall", recall_score(test_y, y_pred, pos_label='GOOD'))
print("Precision", precision_score(test_y, y_pred, pos_label='GOOD'))
print("F1 Score", f1_score(test_y, y_pred, pos_label='GOOD'))
```
